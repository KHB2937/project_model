```python
import pandas as pd
df= pd.read_csv('./송파음식금정카페남음.csv', encoding='utf-8', low_memory=False)
df
```

```python
import re

# Function to clean the text
def clean_text(text):
    # Remove special characters
    text = re.sub(r'\W', ' ', text)
    # Remove numbers
    text = re.sub(r'\d', ' ', text)
    # Remove single characters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    # Substitute multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text, flags=re.I)
    # Convert to lowercase
    text = text.lower()

    return text

# Apply the function to the 'review' column
df['review'] = df['review'].apply(clean_text)

df.head()
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Embedding, LSTM
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from gensim.models import Word2Vec
import numpy as np


# 최대 단어 사전 크기와 최대 문장 길이 설정
MAX_VOCAB_SIZE = 5000
MAX_SENTENCE_LENGTH = 900

# 타겟 레이블 정의
targets = ['데이트', '맛', '서비스', '청결']

# 토크나이저 객체 생성 및 텍스트를 시퀀스로 변환
tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE)
tokenizer.fit_on_texts(df['review'])
sequences = tokenizer.texts_to_sequences(df['review'])

# 시퀀스 패딩
X = pad_sequences(sequences, maxlen=MAX_SENTENCE_LENGTH)

# Word2Vec 모델 학습
sentences = [review.split() for review in df['review']]
word2vec = Word2Vec(sentences, vector_size=64, window=5, min_count=1, workers=4)

# 단어 사전의 모든 단어에 대해 word2vec 벡터를 가져옵니다.
embedding_matrix = np.zeros((MAX_VOCAB_SIZE, 64))
for word, i in tokenizer.word_index.items():
    if i < MAX_VOCAB_SIZE:
        if word in word2vec.wv:
            embedding_matrix[i] = word2vec.wv[word]

# LSTM 모델을 정의합니다.
inputs = Input(shape=(MAX_SENTENCE_LENGTH,))
x = Embedding(MAX_VOCAB_SIZE, 64, weights=[embedding_matrix], input_length=MAX_SENTENCE_LENGTH, trainable=False)(inputs)
x = LSTM(64, dropout=0.2, recurrent_dropout=0.2)(x)

# 각 타겟 레이블에 대한 출력 레이어를 정의합니다.
outputs = [Dense(1, activation='sigmoid')(x) for _ in targets]

# 다중 출력 모델을 정의합니다.
model = Model(inputs=inputs, outputs=outputs)

# 모델을 컴파일합니다.
model.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])

# 각 타겟 레이블에 대해 모델을 학습합니다.
y = [df[target].values for target in targets]
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
y_train = [train_test_split(y_i, test_size=0.2, random_state=42)[0] for y_i in y]
y_test = [train_test_split(y_i, test_size=0.2, random_state=42)[1] for y_i in y]
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Embedding, GRU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from gensim.models import Word2Vec
import numpy as np

# 최대 단어 사전 크기와 최대 문장 길이 설정
MAX_VOCAB_SIZE = 5000
MAX_SENTENCE_LENGTH = 900

# 타겟 레이블 정의
targets = ['데이트', '맛', '서비스', '청결']

# 토크나이저 객체 생성 및 텍스트를 시퀀스로 변환
tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE)
tokenizer.fit_on_texts(df['review'])
sequences = tokenizer.texts_to_sequences(df['review'])

# 시퀀스 패딩
X = pad_sequences(sequences, maxlen=MAX_SENTENCE_LENGTH)

# Word2Vec 모델 학습
sentences = [review.split() for review in df['review']]
word2vec = Word2Vec(sentences, vector_size=64, window=5, min_count=1, workers=4)

# 단어 사전의 모든 단어에 대해 word2vec 벡터를 가져옵니다.
embedding_matrix = np.zeros((MAX_VOCAB_SIZE, 64))
for word, i in tokenizer.word_index.items():
    if i < MAX_VOCAB_SIZE:
        if word in word2vec.wv:
            embedding_matrix[i] = word2vec.wv[word]

# GRU 모델을 정의합니다.
inputs = Input(shape=(MAX_SENTENCE_LENGTH,))
x = Embedding(MAX_VOCAB_SIZE, 64, weights=[embedding_matrix], input_length=MAX_SENTENCE_LENGTH, trainable=False)(inputs)
x = GRU(64, dropout=0.2, recurrent_dropout=0.2)(x)

# 각 타겟 레이블에 대한 출력 레이어를 정의합니다.
outputs = [Dense(1, activation='sigmoid')(x) for _ in targets]

# 다중 출력 모델을 정의합니다.
model = Model(inputs=inputs, outputs=outputs)

# 모델을 컴파일합니다.
model.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])

# 각 타겟 레이블에 대해 모델을 학습합니다.
y = [df[target].values for target in targets]
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
y_train = [train_test_split(y_i, test_size=0.2, random_state=42)[0] for y_i in y]
y_test = [train_test_split(y_i, test_size=0.2, random_state=42)[1] for y_i in y]
model.fit(X_train, y_train, epochs=10, batch_size=32)
```
